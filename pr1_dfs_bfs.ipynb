```python
# DFS using Stack with Random Traversal

import random

graph = {
    0: [1, 2],
    1: [0, 3, 4],
    2: [0, 5],
    3: [1],
    4: [1],
    5: [2]
}

def dfs(start):
    visited = set()
    stack = [start]

    while stack:
        node = stack.pop()

        if node not in visited:
            print(node, end=" ")
            visited.add(node)

            # Randomize neighbor order
            neighbors = random.sample(graph[node], len(graph[node]))

            for neighbor in neighbors:
                if neighbor not in visited:
                    stack.append(neighbor)

print("DFS Traversal:")
dfs(0)
```

```python
# Time Complexity:O(V+E)
# Space Complexity:O(V)
```

```python
from collections import deque

graph = {
    0: [1, 2],
    1: [0, 3, 4],
    2: [0, 5],
    3: [1],
    4: [1],
    5: [2]
}

def bfs(start):
    visited = set()
    queue = deque([start])

    visited.add(start)

    while queue:
        node = queue.popleft()

        print(node, end=" ")

        for neighbor in graph[node]:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)

print("BFS Traversal:")
bfs(0)
```

```python
# Time Complexity:O(V+E)
# Space Complexity:O(V)
```
