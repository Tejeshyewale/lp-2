```python
from flask import Flask, render_template, request

app = Flask(__name__)

@app.route('/', methods=['GET', 'POST'])
def home():

    result = ""

    if request.method == 'POST':

        symptom = request.form['symptom'].lower()

        if "fever" in symptom:
            result = "Possible illness: Viral Infection | Consult: General Physician"

        elif "headache" in symptom:
            result = "Possible illness: Migraine | Consult: Neurologist"

        elif "chest pain" in symptom:
            result = "Possible illness: Heart Problem | Consult: Cardiologist"

        elif "skin" in symptom:
            result = "Possible illness: Skin Allergy | Consult: Dermatologist"

        elif "stomach" in symptom:
            result = "Possible illness: Gastric Problem | Consult: Gastroenterologist"

        else:
            result = "Sorry, symptom not found. Please consult hospital."

    return render_template('index.html', result=result)

if __name__ == '__main__':
    app.run(debug=True)
```
