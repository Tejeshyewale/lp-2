```python
# Selection Sort

arr = [64, 25, 12, 22, 11]

n = len(arr)

for i in range(n):

    min_index = i

    for j in range(i + 1, n):

        if arr[j] < arr[min_index]:
            min_index = j

    # Swap elements
    arr[i], arr[min_index] = arr[min_index], arr[i]

print("Sorted Array:")

for i in arr:
    print(i, end=" ")
```

```python
# Dijkstra Algorithm

graph = {
    'A': {'B': 1, 'C': 4},
    'B': {'C': 2, 'D': 5},
    'C': {'D': 1},
    'D': {}
}

distance = {}
visited = []

# Initialize distances
for node in graph:
    distance[node] = float('inf')

distance['A'] = 0

while len(visited) < len(graph):

    min_node = None

    for node in graph:

        if node not in visited:

            if min_node is None or distance[node] < distance[min_node]:
                min_node = node

    visited.append(min_node)

    for neighbor in graph[min_node]:

        new_distance = distance[min_node] + graph[min_node][neighbor]

        if new_distance < distance[neighbor]:

            distance[neighbor] = new_distance

print("Shortest Distances:")

for node in distance:
    print(node, "=", distance[node])
```

```python


jobs = [
    ('J1', 2, 100),
    ('J2', 1, 50),
    ('J3', 2, 10),
    ('J4', 1, 20)
]

# Sort jobs by profit in descending order
jobs.sort(key=lambda x: x[2], reverse=True)

result = []

time_slots = [False] * 3

for job in jobs:

    job_id = job[0]
    deadline = job[1]
    profit = job[2]

    for i in range(deadline - 1, -1, -1):

        if time_slots[i] == False:

            time_slots[i] = True
            result.append(job_id)
            break

print("Selected Jobs:")

for job in result:
    print(job)
```

```python
# Prim's Minimum Spanning Tree Algorithm

import sys

graph = [
    [0, 2, 0, 6, 0],
    [2, 0, 3, 8, 5],
    [0, 3, 0, 0, 7],
    [6, 8, 0, 0, 9],
    [0, 5, 7, 9, 0]
]

vertices = 5

selected = [False] * vertices

selected[0] = True

print("Edge : Weight")

for k in range(vertices - 1):

    minimum = sys.maxsize
    x = 0
    y = 0

    for i in range(vertices):

        if selected[i]:

            for j in range(vertices):

                if not selected[j] and graph[i][j]:

                    if minimum > graph[i][j]:

                        minimum = graph[i][j]
                        x = i
                        y = j

    print(x, "-", y, ":", graph[x][y])

    selected[y] = True
```

```python
# Kruskal's Algorithm

edges = [
    (1, 'A', 'B'),
    (3, 'B', 'C'),
    (2, 'A', 'C'),
    (4, 'C', 'D')
]

# Sort edges by weight
edges.sort()

print("Edges in MST:")

for edge in edges:
    print(edge[1], "-", edge[2], ":", edge[0])
```

```python
# Dijkstra Algorithm

graph = {
    'A': {'B': 1, 'C': 4},
    'B': {'A': 1, 'C': 2, 'D': 5},
    'C': {'A': 4, 'B': 2, 'D': 1},
    'D': {'B': 5, 'C': 1}
}

distance = {}
visited = []

for node in graph:
    distance[node] = float('inf')

distance['A'] = 0

while len(visited) < len(graph):

    min_node = None

    for node in graph:

        if node not in visited:

            if min_node is None or distance[node] < distance[min_node]:
                min_node = node

    visited.append(min_node)

    for neighbor in graph[min_node]:

        new_distance = distance[min_node] + graph[min_node][neighbor]

        if new_distance < distance[neighbor]:

            distance[neighbor] = new_distance

print("Shortest Distances:")

for node in distance:
    print(node, "=", distance[node])
```
