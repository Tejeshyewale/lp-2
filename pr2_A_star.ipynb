```python
from queue import PriorityQueue

goal = [[1,2,3],
        [4,5,6],
        [7,8,0]]

# Heuristic Function
def heuristic(state):
    count = 0

    for i in range(3):
        for j in range(3):
            if state[i][j] != 0 and state[i][j] != goal[i][j]:
                count += 1

    return count


# Find blank space
def find_blank(state):
    for i in range(3):
        for j in range(3):
            if state[i][j] == 0:
                return i, j


# Generate new states
def moves(state):

    x, y = find_blank(state)

    possible_moves = [(-1,0),(1,0),(0,-1),(0,1)]

    new_states = []

    for dx, dy in possible_moves:

        nx, ny = x + dx, y + dy

        if 0 <= nx < 3 and 0 <= ny < 3:

            temp = [row[:] for row in state]

            temp[x][y], temp[nx][ny] = temp[nx][ny], temp[x][y]

            new_states.append(temp)

    return new_states


# A* Algorithm
def astar(start):

    pq = PriorityQueue()

    pq.put((heuristic(start), start))

    visited = []

    while not pq.empty():

        cost, current = pq.get()

        print(current)

        if current == goal:
            print("Goal Reached")
            return

        visited.append(current)

        for next_state in moves(current):

            if next_state not in visited:

                pq.put((heuristic(next_state), next_state))


# Initial State
start = [[1,2,3],
         [4,0,6],
         [7,5,8]]

astar(start)
```
